Need to identify time value (`x` axis) with birth and death time (`y` axis) and remove.

```python
bd = pers_diag["dgms"][0][:-1,:]
pers = bd[:, 1] - bd[:, 0]
plt.scatter(bd[:,0], pers)
plt.xlabel("Birth Time")
plt.ylabel("Persistence")
plt.show()
```

```python
pers = np.c_[bd, bd[:,1]-bd[:,0]]
```

```python
pers
```

```python
small_pers = pers[pers[:,2] < 0.7, :]
```

```python
small_pers
```

```python
# cols: time index, amp, birth true, death true
A = np.c_[tidx, tst[:,m], np.zeros(tidx.size*2).reshape(-1,2)]
```

```python
A.shape
```

```python
for n, i in enumerate(A):
    if i[1].round(5) in small_pers[:,0].round(5):
        i[2] += 1
    if i[1].round(5) in small_pers[:,1].round(5):
        i[3] += 1
```

```python
A1 = A[A[:,2] != 1, :]
```
