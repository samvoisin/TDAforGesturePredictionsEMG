```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

from data_cube import DataCube
from ssm import SSM
from snf2 import SNF
```

```python
dc = DataCube(
    subjects=["01"],
    gestures=["1", "2", "3", "4"],
    channels=["2", "4", "5", "6", "8"],
    data_grp="parsed"
)
dc.load_data()
dc.rms_smooth(300, 20)
dc.get_max_obs(smooth=True)
```

```python
def euc_dist(i,j,metric,s):
    return abs(i-j)
```

```python
snf_tst = SNF(dc.data_set_smooth["01"]["1_1_1"][:, :6], k=0.5)
snf_tst.normalize_modalities()

# need to calc weights to find knn
snf_tst.calc_weights()
snf_tst.normalize_weights()
# generate and normalize knn graph
snf_tst.calc_knn_weights()
snf_tst.normalize_knn_weights()

snf_tst.network_fusion()

snf_tst.plot_template()
```

```python
tidx = snf_tst.tidx
mods = snf_tst.mods

plt.plot(range(tidx.size), mods[:,0])
plt.plot(range(tidx.size), mods[:,1])
plt.plot(range(tidx.size), mods[:,2])
plt.plot(range(tidx.size), mods[:,3])
plt.plot(range(tidx.size), mods[:,4])
plt.show()
```
